# Univariate Analysis

```python
df_cat= df.select_dtypes(include=object)
df_num= df.select_dtypes(include='number')
print(df_cat.columns)
print(df_num.columns)
```

```python
def univ_category(data,angle):
    print('Analysis for ' + data)
    print('--------------------------------------------------------------------------------')
    print(df_cat[data].describe(), end=' ')
    print('--------------------------------------------------------------------------------')
    if df_cat[data].nunique() < 10:
        ax= sns.countplot(data=df_cat,x=data)
        for bars in ax.containers:
            ax.bar_label(bars,[f'{x:.0f}' for x in bars.datavalues],padding=1.5)
        plt.title(data)
    else:
        ax1= df_cat[data].value_counts().head(10).plot(kind='bar')
        for bars1 in ax1.containers:
            ax1.bar_label(bars1,[f'{x:.0f}' for x in bars1.datavalues],padding=1.5)
        plt.title('Top 10 ' + data)
    plt.xlabel(data)
    plt.xticks(rotation=angle)
    plt.show()
```

```python
for cols in df_cat.columns:
    univ_category(cols,75)
```

```python
def univ_numerical(num_data,nbins):
    print('Analysis for ' + num_data)
    print('-------------------------------------------------------------------------------------------------------------')
    print(df_num[num_data].describe(), end=' ')
    print('----------------------------------------------------------------------------------------------------------------')
    sns.histplot(data=df_num,x=num_data,kde=True,bins=nbins)
    plt.title(num_data)
    plt.xlabel(num_data)
    plt.show()
```

```python
for cols1 in df_num.columns[1:]:
    univ_numerical(cols1,10)
```

# Bivariate analysis

```python
def biv_cat(data_1):    
    for a in data_1.columns[::2]:
            for b in data_1.columns[1::2]:
                plt.figure(figsize=(9,5))
                if (a =='CITY' and  b=='COUNTRY') or (a=='COUNTRY' and b=='CITY') or a=='CUSTOMERNAME' or b=='CUSTOMERNAME' or (a=='STATUS' and b=='CITY'):
                    continue
                else:
                    table=pd.crosstab(index=data_1[a],columns=data_1[b],margins=True)
                    sns.heatmap(data=table,annot=True,fmt='.0f')
                    plt.title(a+' vs '+b)
                plt.show()
```

```python
biv_cat(df_cat)
```

```python
def biv_num(data_2):
    for j in data_2.columns:
            for k in ['QUANTITYORDERED','SALES']:
                if j not in ['STATUS','CITY','CUSTOMERNAME']:
                    plt.figure(figsize=(10,5))
                    group1= df.groupby(j)
                    table1=group1[k].sum()
                    tab1=group1[k].sum().reset_index()
                    best_bar = tab1.loc[tab1[k].idxmax(),j]
                    colour = ['teal' if value== best_bar else 'skyblue' for value in tab1[j]]
                    print(table1)
                    print('----------------------------------------------------------------------------------------------------------------')
                    ax=sns.barplot(data=table1,palette=colour,width=0.9)
                    for bars in ax.containers:
                        ax.bar_label(bars,[f'{x:.2f}' for x in bars.datavalues],padding=1.5)
                    plt.xticks(rotation=70)
                    plt.title(j+' vs '+k)
                    plt.show()
```

```python
biv_num(df_cat)
```

```python
df_corr = df_num.corr()
sns.heatmap(data=df_corr,fmt='.2f',annot=True,cmap='seismic_r')
plt.title('Correlation Matrix')
plt.show()
```

```python
def time_analysis(time,chart):
    sns.set(style='whitegrid')
    plt.figure(figsize=(9,5))
    if chart == 'line':
        group = df.groupby(time)
        time_plot = group['SALES'].sum()
        print(time_plot)
        print('----------------------------------------------------------------------------------------------')
        time_plot.plot(kind=chart)
        plt.title(time)
        plt.xlabel(time)
        plt.show()
        print('----------------------------------------------------------------------------------------------')
    else:
        group = df.groupby(time)
        time_plot = group['SALES'].sum()
        time_plt=group['SALES'].sum().reset_index()
        best_day = time_plt.loc[time_plt['SALES'].idxmax(),time]
        color=['teal' if val== best_day else 'skyblue' for val in time_plt[time] ]
        print(time_plot)
        print('----------------------------------------------------------------------------------------------')
        time_plot.plot(kind=chart,color=color)
        plt.title(time)
        plt.xlabel(time)
        plt.show()
        print('----------------------------------------------------------------------------------------------')
```

```python
for i in df.columns[-4:-2]:
    time_analysis(i,'line')
```

```python
for i in df.columns[-2:]:
    time_analysis(i,'bar')
```

# Multivariate analysis

```python
def multivariate(data_3,legend):
    for j in data_3.columns[:-1]:
            for k in ['QUANTITYORDERED','SALES']:
                if j not in ['STATUS','CITY','CUSTOMERNAME']:
                    plt.figure(figsize=(10,5))
                    group1= df.groupby([j,legend])
                    table1=group1[k].sum()
                    print(table1)
                    print('----------------------------------------------------------------------------------------------------------------')
                    sns.barplot(data=df,x=j, y=k, palette='magma', hue=legend, width=0.9, estimator='sum')
                    plt.xticks(rotation=70)
                    plt.title(j+' vs '+k)
                    plt.show()
```

```python
multivariate(df_cat,'DEALSIZE')
```
